# First steps 4 : Customizing your plot

## Using Themes

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, curdoc
```

```python
output_notebook()
```

```python
# prepare some data
x = [1, 2, 3, 4, 5]
y = [4, 5, 5, 7, 2]
```

```python
# apply theme to current document
curdoc().theme = "dark_minimal"
```

- Bokeh의 테마를 사용하면 plot의 모양을 빠르게 변경할 수 있음
    - 테마는 색상, 글꼴 또는 선 스타일과 같은 미리 정의된 디자인 매개변수 세트
    - 내장 테마를 사용할 수도 있고 사용자가 정의한 테마를 사용할 수 있음
        - 기본 제공 테마를 사용하려면 사용하려는 테마의 이름을 [theme](https://docs.bokeh.org/en/latest/docs/reference/themes.html#module-bokeh.themes) 문서에서 찾아서 사용
        - [사용자 정의 테마](https://docs.bokeh.org/en/latest/docs/user_guide/styling.html#userguide-styling-using-themes-custom)

```python
# create a plot
p = figure(sizing_mode="stretch_width", max_width=500, height=250)
```

```python
# add a renderer
p.line(x, y)
```

```python
# show the results
show(p)
```
